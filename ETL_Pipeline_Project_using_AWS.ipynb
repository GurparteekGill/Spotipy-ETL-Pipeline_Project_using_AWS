# Instead of extracting the data of one single album we will exract the data of all the 50 albums using Loop

# 1.

```python
album_list=[]

for row in data['items']:
    album_id=row['track']['album']['id']
    album_name=row['track']['album']['name']
    album_release_date=row['track']['album']['release_date']
    album_total_tracks=row['track']['album']['total_tracks']
    album_url = row['track']['album']['external_urls']['spotify']
    album_element= {'album_id':album_id,'name':album_name,'release_date': album_release_date,'toal_tracks':album_total_tracks
                   ,'url':album_url}
    
    album_list.append(album_element)
    #print(album_element)               
```

```python
# What if i want to get the artist_name , but here you can see that one song is having  total 2 artist 
# Note that One song can have one than one artist . Its a type of One to many relationship
# This type is know as Nested JSON.
```

# 2.  One-TO-Many situation

```python
data['items'][9]['track']['artists']
```

```python
#So we will trackle the situatuin in this way
```

```python


artist_list=[]

for row in data['items']:
    for key,value in row.items (): # bECAUSE WE have the dictinary inside 
        if key == 'track':
            #print(value['artists'])
            for artist in value['artists']:
                #print(artist['id'])
                artist_dict=  {'artist_id':artist['id'],'artist_name':artist['name'],'externl_url': artist['href']}
                artist_list.append(artist_dict)
```

```python
artist_list
```

# 3. Same Thing I can do For the Song_list

```python
song_list=[]

for row in data['items']:
    song_id=row['track']['id']
    song_name=row['track']['name']
    song_release_date=row['track']['album']['release_date']
    song_total_tracks=row['track']['album']['total_tracks']
    song_url = row['track']['external_urls']['spotify']
    song_duration = row['track']['duration_ms']
    song_added = row['added_at']
    
    song_elements= {'song_id':song_id,'song_name':song_name,'song_release_date': song_release_date,'song_toal_tracks':song_total_tracks
                   ,'song_url':song_url,'song_added':song_added,'song_duration':song_duration}
    
    song_list.append(song_elements)
    
    
    
```
